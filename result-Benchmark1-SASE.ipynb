### - import ebeam file, static seed laser.

```python
from functions import GenesisOut
import matplotlib.pyplot as plt
#filepath = './benchmark/Benchmark1-SASE/Benchmark.out.h5'
filepath = './benchmark/Benchmark1-SASE/Benchmark.out.h5'
out = GenesisOut.GenesisOut(filepath)
```

```python
out.show_info('Beam')
```

```python
out.show_info('Field')
```

```python
plt.plot(out.zplot, out('Field', 'xsize'), label='x')
plt.plot(out.zplot, out('Field', 'ysize'), label='y')
plt.legend()
plt.xlabel('z [m]')
plt.ylabel('field size')
```

```python
plt.plot(out.zplot, out('Beam', 'xsize'), label='x')
plt.plot(out.zplot, out('Beam', 'ysize'), label='y')
plt.legend()
plt.xlabel('z [m]')
plt.ylabel('beam size')
```

```python
plt.plot(out.zplot, out('intensity-farfield'), label='far field')
plt.plot(out.zplot, out('intensity-nearfield'), label='near field')
plt.xlabel('z [m]')
plt.ylabel('intensity')
plt.show()
```

```python
plt.plot(out.zplot, out('Field', 'power'))
plt.xlabel('z [m]')
plt.ylabel(' P [W]')
plt.title('power')
plt.show()
```
